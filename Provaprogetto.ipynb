```python
import os
import zipfile
import gdown
import shutil

# Monta Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Percorso principale del dataset DOTA in Google Drive
base_path = '/content/drive/MyDrive/DOTA'

# Crea la cartella principale se non esiste
if not os.path.exists(base_path):
    os.makedirs(base_path)

### Funzione per scaricare ed estrarre i file .zip in una cartella specifica ###
def download_and_extract(folder_id, dest_folder):
    # Scarica l'intera cartella usando gdown
    gdown.download_folder(f"https://drive.google.com/drive/folders/{folder_id}", output=dest_folder, quiet=False)

    # Estrai tutti i file .zip nella cartella di destinazione
    for file_name in os.listdir(dest_folder):
        if file_name.endswith('.zip'):
            file_path = os.path.join(dest_folder, file_name)
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                zip_ref.extractall(dest_folder)
            print(f"{file_name} estratto con successo.")
            # Rimuovi il file zip dopo l'estrazione per risparmiare spazio
            os.remove(file_path)

### 1. Organizza il Validation Set ###
# ID della cartella di Google Drive per il Validation Set
validation_folder_id = '1n5w45suVOyaqY84hltJhIZdtVFD9B224'
validation_path = os.path.join(base_path, 'validation')

# Crea la cartella di destinazione per il Validation Set
if not os.path.exists(validation_path):
    os.makedirs(validation_path)

# Scarica ed estrai i file di validation
download_and_extract(validation_folder_id, validation_path)

# Sposta le immagini in una sottocartella chiamata 'images' (opzionale)
validation_images_folder = os.path.join(validation_path, 'images')
if not os.path.exists(validation_images_folder):
    os.makedirs(validation_images_folder)

for item in os.listdir(validation_path):
    item_path = os.path.join(validation_path, item)
    if item_path.endswith('.png') or item_path.endswith('.jpg'):
        shutil.move(item_path, validation_images_folder)

print("Validation set organizzato correttamente.")


### 2. Organizza il Test Set ###
# ID della cartella di Google Drive per il Test Set
test_folder_id = '1mYOf5USMGNcJRPcvRVJVV1uHEalG5RPl'
test_path = os.path.join(base_path, 'test')

# Crea la cartella di destinazione per il Test Set
if not os.path.exists(test_path):
    os.makedirs(test_path)

# Scarica ed estrai i file di test
download_and_extract(test_folder_id, test_path)

# Sposta le immagini in una sottocartella chiamata 'images' (opzionale)
test_images_folder = os.path.join(test_path, 'images')
if not os.path.exists(test_images_folder):
    os.makedirs(test_images_folder)

for item in os.listdir(test_path):
    item_path = os.path.join(test_path, item)
    if item_path.endswith('.png') or item_path.endswith('.jpg'):
        shutil.move(item_path, test_images_folder)

print("Test set organizzato correttamente.")
```

```python
import os
import random
import matplotlib.pyplot as plt
import cv2

# Percorso alla cartella delle immagini di training
training_images_path = '/content/drive/MyDrive/DOTA/training/images'

# Lista di tutte le immagini nella cartella di training
all_images = [f for f in os.listdir(training_images_path) if f.endswith('.png') or f.endswith('.jpg')]

# Numero di immagini da visualizzare
num_images = 5
sample_images = random.sample(all_images, num_images)

# Mostra le immagini selezionate
plt.figure(figsize=(15, 5))
for i, image_name in enumerate(sample_images):
    image_path = os.path.join(training_images_path, image_name)
    # Carica l'immagine con OpenCV
    image = cv2.imread(image_path)
    # Converte l'immagine da BGR (OpenCV) a RGB (Matplotlib)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    # Visualizza l'immagine
    plt.subplot(1, num_images, i + 1)
    plt.imshow(image)
    plt.title(image_name)
    plt.axis('off')
plt.show()
```
